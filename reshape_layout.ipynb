```python
import pandas as pd
titanic=pd.read_csv('https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/titanic.csv')
air_quality=pd.read_csv('https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/air_quality_long.csv')
```

```python
##I want to sort the Titanic data according to the age of the passengers.->df.sort_values(by='')
titanic.sort_values(by="Age").head()
```

```python
##I want to sort the Titanic data according to the cabin class and age in descending order.
titanic.sort_values(by=['Age','Pclass'],ascending=False).head()
```

```python

air_quality.head()
```

```python
# filter for no2 data only
no2=air_quality[air_quality['parameter']=='no2']
no2.head()
```

```python
##I want the values for the three stations as separate columns next to each other
## pivot() function is purely reshaping of the data: a single value for each index/column combination is required.
no2_subset = no2.sort_index().groupby(["location"]).head(2)
no2_subset.pivot(columns="location", values="value")
```

```python
##I want the mean concentrations for no2 and pm25 in each of the stations in table form
##index에는 열 정보가 들어가고 columns에 행 정보가 들어간다.
air_quality.pivot_table(values='value',index='location',columns='parameter',aggfunc='mean')
```

```python
##When interested in summary columns for each variable separately as well, put the margin parameter to True:
air_quality.pivot_table(values='value',index='location',columns='parameter',aggfunc='mean',margins=True)
```

```python
no2_pivoted = no2.pivot(columns="location", values="value").reset_index()
no2_pivoted.head()
```

```python
no_2 = no2_pivoted.melt(id_vars="date.utc")
no_2.head()
```
